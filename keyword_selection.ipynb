```python
from trendspy import Trends

tr_ng = Trends(language="en-US", tzs=60, use_enitity_names=True, request_delay=60, proxy=None)               # volume-specific client

timeframe = "2011-01-01 2023-12-31"

# Interest Over Time for one keyword (topic) at a time for NG based on start and end date
def trend_data_one_keyword(keyword, geo='NG', timeframe=timeframe):

    df = tr_ng.interest_over_time(keyword)
    df.to_csv(f"{keyword}.csv", index=False)

    return df
```
